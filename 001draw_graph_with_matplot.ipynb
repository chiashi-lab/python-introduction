```python
basic_vec1 = np.array([1,2])
basic_vec2 = np.array([-2,1])

for i in range(10):
    plt.scatter(i*basic_vec1[0], i*basic_vec1[1], color='red')
    plt.scatter(i*basic_vec2[0], i*basic_vec2[1], color='blue')

plt.show()
```

```python
for i in range(10):
    for j in range(10):
        plt.scatter(i*basic_vec1[0] + j*basic_vec2[0], i*basic_vec1[1] + j*basic_vec2[1], color='green')

plt.show()
```
